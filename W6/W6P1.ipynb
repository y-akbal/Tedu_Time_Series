```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
def return_random_walk(n :int, mu: float =  0.1, p0 :float = 0) -> list:
    A = []
    A.append(p0)
    for i in range(n):
      A.append(A[-1] + mu + 10*np.random.randn())
    return A
```

```python
rw = return_random_walk(1000)
```

```python
plt.plot(rw)
```

```python
def return_time_series_lin_trend(n :int, mu: float) -> list:
    return [mu*i+10*np.random.randn() for i in range(n)]
```

```python
lt_s = return_time_series_lin_trend(1000, mu = 0.1)
```

```python
plt.plot(lt_s)
```
