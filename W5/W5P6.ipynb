```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
def function(phi0 :float, phi1: float, phi2: float, length :int = 10, initial_pt:tuple = (0,0)):   ### remember this function from the first weeks?
    A = []
    for i in initial_pt: ### initial points are appended to the array
        A.append(i)
    f = lambda x1, x2 : phi0 + phi1*x1 + phi2*x2+ 5*np.random.randn()  ## this is a lambda function
    for i in range(length-1):
        A.append(f(A[-1], A[-2]))  ### appending the elements in an autoregressive way
    return A
```

```python
A = function(1, 26/25, -1/25, 250)  ### guess what sort of thing we have here?
plt.plot(A)
plt.title("I am a random walk!")
```

```python
fig, ax = plt.subplots(1, 8, figsize = (25,5))

for i,a in enumerate(ax):
    
    a.plot(function(1, 26/25, -1/25, 550))
    a.title.set_text(f'I am {i}th freak')
plt.legend()
plt.show()
```
