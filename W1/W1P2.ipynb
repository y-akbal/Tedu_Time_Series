#How to create a random vector?

```python
dir(np.random)  ### below you will see distributions!
```

```python
np.random.poisson(2,size = (15,5))   ### a random sample from poisson distribution with mean 2.
```

```python
np.random.randn(10,10)   #### A sample from standard normal distribution 
```

```python
np.random.normal()
```

```python
np.random.normal(0, 2, size = (10,10)) #### fixing variable and variance
```

```python
np.random.chisquare(2, size = (10,10))  ### sample from chi^2 distribution with degrees of freedom 2.
```

#Numpy arrays of the same size can be multiplied as matrices


```python
B = np.random.randn(10,10)
C = np.random.randn(10,10)
T = np.dot(B,C)  #### as a matrix B*C
print(T)
```

```python
np.array_equal(np.dot(B,C), np.matmul(B,C))  ### this arrays are not equal because B*C is component wise multiplication
```

```python
A = np.random.randn(10,1)
B = np.ones(shape = (10,1))
A*B
```
