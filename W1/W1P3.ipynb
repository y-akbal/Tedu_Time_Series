```python
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np
```

#Sketch simple function

```python
f = lambda x : x**3 -x + 1
```

```python
r = np.arange(0,20,0.001)
```

```python
plt.plot(r, f(r), label = "function")
plt.legend()
plt.show()
```

# Multiple functions

```python
g = lambda x : x**3 +x+1
```

```python
plt.plot(r, f(r), label = "function1")
plt.plot(r, g(r), label = "function2")
plt.legend()
plt.show()
```

#Scatter Plot

```python
L = np.random.randn(1000)
plt.plot(L, label = "Time Series: White noise", color = "red")
plt.legend()
plt.show()
```

#Histogram Plotting

```python
plt.hist(np.random.randn(1000000), bins = 100, density =  True, label = "standard")
plt.hist(1.5*np.random.randn(1000000)+1, bins = 109, density =  True, color = "red", label = "shifted")
plt.legend()
plt.show()
```

#Multiple canvas' all in once!

```python
X = range(1000)
Y = np.random.randn(1000)
fig, ax = plt.subplots(5,2, squeeze = False)
for ax_1 in ax:   #### ax is a numpy array BTW!
  for i,ax_2 in enumerate(ax_1):
    ax_2.plot(X,Y, label = f"label {i}")
plt.legend()
plt.show()
```
