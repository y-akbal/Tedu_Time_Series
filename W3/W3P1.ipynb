Below we create a function returning correlation values of two random vectors that are sampled from standard normal dist.

```python
def f(length: int, length_of_array: int = 100):
    L = []
    for _ in range(length):
        x = np.random.randn(length_of_array)
        y = np.random.randn(length_of_array)
        L.append(scipy.stats.pearsonr(x,y)[0])
    return L
```

```python
l = f(100000, 1000)
```

```python
plt.hist(l, bins = 100)
plt.show()   #### Does it show that this data is normally distributed?
```

```python
np.mean(l)
```

```python
np.std(l, ddof = 1)
```
