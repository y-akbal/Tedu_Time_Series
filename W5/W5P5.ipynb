```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
def function(length_:int  = 1000, in_pt:int = 0, **kwargs) -> list:
  pt = in_pt
  L = []
  L.append(pt)
  for i in range(length_):
    pt += np.random.normal(**kwargs)
    L.append(pt)
  return L
```

```python
plt.plot(function(12312, in_pt = 2, loc = 0, scale = 0.0001))
plt.title("I am the ugliest random walk!")
plt.show()
```

```python
def Random_walk_with_drift(length_:int  = 1000, in_pt:int = 0, drift:float = 0,**kwargs) -> list:
  pt = in_pt
  L = []
  L.append(pt)
  for i in range(length_):
    pt += drift + np.random.normal(**kwargs)
    L.append(pt)
  return L
```

```python
drift = 0.05*np.random.randn()
plt.plot(Random_walk_with_drift(12312, 2, drift))
plt.title(f"I am the second ugliest random walk, and I have a drift {drift:.02f}")
plt.show()
```

```python
drift = 0.005*np.random.randn()
rw = Random_walk_with_drift(12312, 2, drift)
drift_lin = [drift*i for i,_ in enumerate(rw)]
plt.figure(figsize=(10,5))
plt.plot(rw, label = "The random dude")
plt.plot(drift_lin, label = "Slope")
plt.title(f"I am the second ugliest random walk, and I have a drift {drift:.03f}")
plt.legend()
plt.show()
```
