```python
data = pd.read_csv("tsfinal02.csv")
```

```python
def sliding_window(data_, lag_size = 50):  
  N = len(data_)
  data = np.array(data_)
  X = []
  y = []
  for i in range(N-lag_size-1):
    X.append(data[i:i+lag_size])
    y.append(data[i+lag_size])
  return np.array(X), np.array(y)
```
