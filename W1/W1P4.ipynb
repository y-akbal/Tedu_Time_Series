#An important function to be used a lot to_dataframe.py. It converts a txt file to pandas data frame

```python

def to_dataframe(text):
    """
    
    Parameters
    ----------
    text : this dude is a text file.

    Returns
    -------
    Panda dataframe

    """    
    L = []
    headers = []
    with open(text) as file:
        T = file.readlines()
        for i, t in enumerate(T):
            if i == 0:
                for header  in t[:-1].split():
                  headers.append(header)      
            else:
                L.append(t[0:-1].split())
    L = pd.DataFrame(L)
    
    L = L.apply(pd.to_numeric)
    L.columns = headers

    return L
```
