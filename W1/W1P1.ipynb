# List comprehension

```python
A = []
def function(A:list, n:int)-> list:
    for i in range(n):
        A.append(i)
    return A
    
```

```python
function(A, 5)
```

```python
A = [i for i in range(5)]   ### this is called list comprehension
```

```python
A = [i*j for i in range(5) for j in range(5)]
```

```python
A
```

```python
len(A)
```

```python
A = set(i*j for i in range(5) for j in range(5))
```

```python
len(A)
```

# enumerate, zip

```python
A = ["apple", "banana", "orange"]
for i, _ in enumerate(A):
    print(i, _)
```

```python
for i in range(len(A)):
  print(i, A[i])
```

```python
B = ["horse", "bear", "wasp"]
for _, __ in zip(A,B):
    print(_, __)
```

```python
type(zip(A,B))
```

```python
B = ["horse", "bear", "wasp", "snake"]   ###shorter one is important
for _, __ in zip(A,B):
    print(_, __)
```

# ***Exception Handling***

```python
A = [1,2,3,4]
try:
  A[5]  #### trying to do this 
except Exception as a:
  print("sd") 
  print(f"Exception here is {a}")
finally:
   print("I am done here leaving")
```
