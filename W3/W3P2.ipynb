```python
import seaborn as sns
from matplotlib import pyplot as plt
import numpy as np
```

```python
phi0 = np.random.randn()
phi1 = np.random.randn()
initial_pt = 0
```

```python
def function(phi0 :float, phi1: float, length = 10 , initial_pt = 0):
    A = [] #start with an empty list
    A.append(initial_pt)
    f = lambda x : phi0 + phi1*x + 0.5*np.random.randn()
    for i in range(length-1):
        A.append(f(A[-1]))
    return [i for i, _ in enumerate(A)] ,A
```

```python
phi0 = np.random.randn()
phi1 = np.random.randn()
ts = function(phi0, phi1 ,length = 1000, initial_pt = 2)
plt.plot(ts[0], ts[1], label =f"phi_1 = {phi1:.2f} and phi_0 = {phi0:.2f}")
plt.legend()
plt.show()
```
