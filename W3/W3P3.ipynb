```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
```

```python
#np.random.seed(5)  ### fixing the random seed to get the same result always adjust this dude as you like!
phi_0 = np.random.randn()
phi_1 = np.random.randn()
phi_2 = np.random.randn()
initial_pt = np.random.randn(), np.random.randn()
```

```python
def function(phi0 :float, phi1: float, phi2: float, length :int = 10, initial_pt:tuple = (0,0)):
    A = []
    for i in initial_pt: ### initial points are appended to the array
        A.append(i)
    f = lambda x1, x2 : phi0 + phi1*x1 + phi2*x2+ 0.5*np.random.randn()  ## this is a lambda function
    for i in range(length-1):
        A.append(f(A[-1], A[-2]))  ### appending the elements in an autoregressive way
    return [i for i, _ in enumerate(A)] ,A
```

#Below experiment with some $\phi_0, \phi_1$ and $\phi_2$, which resulting time series appear to be stationary?

```python
ts = function(phi_0, phi_1, phi_2 ,length = 250, initial_pt = initial_pt)
plt.plot(ts[0], ts[1], label =f"phi_0 = {phi_0:.2f} and phi_1 = {phi_1:.2f}+ phi_2 = {phi_2:.2f}")
plt.legend()
plt.show()
```
