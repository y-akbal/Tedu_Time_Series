```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
def return_time_series_lin_trend(n :int, mu: float) -> list:
    return [mu*i+np.random.randn() for i in range(n)]
```

```python
lt_s = return_time_series_lin_trend(1000, mu = 0.2)
```

```python
plt.plot(lt_s)   ### you see there is an obvious deterministic trend, to make sure this let's apply KPSS test
```

```python
#statsmodels.tsa.stattools.kpss
import statsmodels
from statsmodels.tsa.stattools import kpss
```

```python
kpss(lt_s, regression = "ct")
```
