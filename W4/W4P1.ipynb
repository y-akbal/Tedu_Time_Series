```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
A = 0.5*np.random.randn(1000)
theta_1 = np.random.randn()
c_0 = np.random.randn()
```

```python
def create_ma1(A, theta_1, c_0 = 0):
  L = []
  L.append(c_0)
  f = lambda at, at1 : c_0+at - theta_1*at1
  for i, _ in enumerate(A):
    try:  ###we use exception blocks to pertain interpreter thrown an error!
      L.append(f(A[i], A[i+1]))
    except Exception:
      break
  return L
```

```python
ma1 = create_ma1(A, 2, c_0 = 0)
plt.figure(figsize = (15,3))
plt.plot(ma1, label = "MA(1) Model")
plt.legend(fontsize = 10)
plt.show()
```

```python
np.var(A)
```

```python
np.var(ma1)
```

See these two guys are almost the same, above we have the variance of the series. Below we have the variance coming from the formula!

```python
(1+(-1)**2)*np.var(A)
```
